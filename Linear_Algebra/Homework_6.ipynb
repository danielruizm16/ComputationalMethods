$$
A \;=\; D \;+\; L \;+\; U,
$$

where:

- $D$ is the diagonal matrix of $A$,
- $L$ is the strictly lower triangular part,
- $U$ is the strictly upper triangular part.

The Gauss-Seidel algorithm rearranges the system $Ax=b$ as

$$
x \;=\; (D+L)^{-1}\bigl(b \;-\; Ux\bigr),

$$
x_i^{(k+1)}
= \frac{1}{a_{ii}}

\qquad i=1,\dots,n.
$$

The SOR method is an extension of the Gauss-Seidel method that uses a relaxation factor $\omega$ to accelerate convergence. It decomposes the matrix A as:

$$
A = D + L + U,
$$
where:
- $D$: Diagonal matrix of $A$
- $L$: Strictly lower triangular matrix
- $U$: Strictly upper triangular matrix

The SOR iteration formula is:

$$
x^{(k+1)} = (D + \omega L)^{-1}[\omega b - (\omega U + (\omega - 1)D)x^{(k)}]

$$

$$
