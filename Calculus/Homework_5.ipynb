Boole's rule and Weddle's rule are numerical integration methods that use higher-order polynomial approximations. Write Python functions to implement both rules and compare their performance.

**Boole's Rule**
Uses 5 points (4th degree polynomial):
$$
\int_a^b f(x)dx \approx \frac{2h}{45}[7f(x_0) + 32f(x_1) + 12f(x_2) + 32f(x_3) + 7f(x_4)]
$$
where $h = \frac{b-a}{4}$ and $x_i = a + i\cdot h$


**Weddle's Rule**
Uses 7 points (6th degree polynomial):
$$
\int_a^b f(x)dx \approx \frac{3h}{10}[f(x_0) + 5f(x_1) + f(x_2) + 6f(x_3) + f(x_4) + 5f(x_5) + f(x_6)]
$$
where $h = \frac{b-a}{4}$ and $x_i = a + i\cdot h$
